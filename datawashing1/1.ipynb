```python
import pandas as pd
from streamlit import title
from ydata_profiling import ProfileReport

df = pd.read_csv("data.csv", encoding="latin-1")
```

```python
print(df.isnull().sum()) #每列缺失值数
print(df.isnull().mean()) #缺失值比例
df.info() #查找非空计数及data type
```

```python
df['Amount'] = df['Quantity'] * df['UnitPrice']
#处理退货金额为负的交易
df['TransactionType'] = df['Amount'].apply(lambda x: 'Refund' if x < 0 else 'Sale') #新建类型
df['Amount_Abs'] = df['Amount'].abs() #金额转为正数
print(df.head())
```

```python
df = df[df['Quantity'] > 0] #去除退货/负数量
df =df[df['UnitPrice'] > 0] #去除单价异常记录
df.dropna(inplace=True)

profile = ProfileReport(df, title='Sale Data Report')
profile.to_file('index.html')
```

```python
df.to_csv('cleaned_sales.csv', index=False)
```
